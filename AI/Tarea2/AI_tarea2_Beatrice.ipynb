```python
from google.colab import drive
drive.mount('/content/drive')

import os
import torch
from torchvision import transforms, models
import numpy as np
from sklearn.metrics import average_precision_score, precision_recall_curve
import matplotlib.pyplot as plt
import pandas as pd
import skimage.io as io
import skimage.transform as transform
import clip

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Define datasets
datasets = {
    "Simple1K": "/content/drive/MyDrive/Datasets/simple1K",
    "VOC-Pascal": "/content/drive/MyDrive/Datasets/VOC_val",
    "Paris": "/content/drive/MyDrive/Datasets/Paris_val"
}

# Preprocessing for different models
preprocess_resnet = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

preprocess_clip = None

# Load models
def load_resnet34():
    weights = models.ResNet34_Weights.IMAGENET1K_V1
    model = models.resnet34(weights=weights).to(device)
    model.fc = torch.nn.Identity()
    model.eval()
    return model, preprocess_resnet, 512

def load_dinov2():
    model = torch.hub.load('facebookresearch/dinov2', 'dinov2_vits14').to(device)
    model.eval()
    return model, preprocess_resnet, 384

def load_clip():
    global preprocess_clip
    model, preprocess_clip = clip.load("ViT-B/32", device=device)
    model.eval()
    return model, preprocess_clip, 512

modelss = {
    "resnet34": load_resnet34,
    "dinov2": load_dinov2,
    "clip": load_clip
}

# Load dataset
def load_dataset(dataset_path):
    list_file = os.path.join(dataset_path, "list_of_images.txt")
    df = pd.read_csv(list_file, sep="\t", names=["image_name", "class_id"])
    df["image_name"] = df["image_name"].str.split().str[0]
    image_paths = [os.path.join(dataset_path, "images", img) for img in df["image_name"]]
    labels = df["class_id"].values
    print(f"Dataset: {dataset_path}, Total images: {len(image_paths)}")
    print(f"First 5 image paths: {image_paths[:5]}")
    return image_paths, labels

# Extract features
def extract_features(model, preprocess, image_path, feature_size):
    try:
        image = Image.open(image_path).convert("RGB")
        input_tensor = preprocess(image).unsqueeze(0).to(device)
        with torch.no_grad():
            if "clip" in model.__class__.__name__.lower():
                features = model.encode_image(input_tensor).squeeze().cpu().numpy()
            else:
                features = model(input_tensor).squeeze().cpu().numpy()
        max_abs = np.max(np.abs(features))
        if not np.all(np.isfinite(features)):
            print(f"Warning: Raw features for {image_path} contain inf/NaN")
        if max_abs > 1e10:
            print(f"Warning: Raw features for {image_path} are too large: max={max_abs}")
        norm = np.linalg.norm(features)
        if norm == 0 or not np.all(np.isfinite(features)) or max_abs > 1e10:
            print(f"Warning: Invalid features for {image_path}, norm: {norm}, max_abs: {max_abs}")
            fallback = np.zeros(feature_size) + 1e-5
            return fallback / np.linalg.norm(fallback)
        return features / norm
    except Exception as e:
        print(f"Error processing {image_path}: {e}")
        return np.zeros(feature_size) + 1e-5

# Visualize retrieval results
def visualize_results(query_idx, top_k, bottom_k, image_paths, labels, title_prefix):
    fig, ax = plt.subplots(2, 6, figsize=(15, 5))
    query_img = io.imread(image_paths[query_idx])
    query_img = transform.resize(query_img, (64, 64))
    ax[0, 0].imshow(query_img)
    ax[0, 0].set_axis_off()
    ax[0, 0].set_title(f"Query: {labels[query_idx]}")

    for i, idx in enumerate(top_k[:5], 1):
        img = io.imread(image_paths[idx])
        img = transform.resize(img, (64, 64))
        ax[0, i].imshow(img)
        ax[0, i].set_axis_off()
        ax[0, i].set_title(f"Class: {labels[idx]}")

    for i, idx in enumerate(bottom_k[:5], 1):
        img = io.imread(image_paths[idx])
        img = transform.resize(img, (64, 64))
        ax[1, i].imshow(img)
        ax[1, i].set_axis_off()
        ax[1, i].set_title(f"Class: {labels[idx]}")

    plt.suptitle(f"{title_prefix} Retrieval Results")
    plt.tight_layout()
    plt.show()

# Evaluate
def evaluate_model(features, labels, image_paths, model_name):
    norm2 = np.linalg.norm(features, ord=2, axis=1, keepdims=True)
    print(f"Feature norms (first 5): {norm2.flatten()[:5]}")
    norm2[norm2 == 0] = 1.0
    feats_n = features / norm2
    sim = feats_n @ feats_n.T
    sim_max = np.max(sim)
    print(f"Raw max similarity: {sim_max}")
    sim = np.clip(sim, -1.0, 1.0)
    if not np.all(np.isfinite(sim)):
        print(f"Warning: Similarity matrix contains inf/NaN after clipping")
        sim[~np.isfinite(sim)] = -1.0
    sim_idx = np.argsort(-sim, axis=1)

    mAP_scores = []
    for i in range(len(image_paths)):
        query_label = labels[i]
        similarities = sim[i].copy()
        similarities[i] = -1.0
        y_true = (labels == query_label).astype(int)
        y_true[i] = 0
        y_score = similarities
        y_score_max = np.max(np.abs(y_score))
        if np.all(np.isfinite(y_score)) and y_score_max <= 1e10:
            try:
                ap = average_precision_score(y_true, y_score)
                mAP_scores.append(ap)
            except ValueError as e:
                print(f"Warning: AP computation failed for {image_paths[i]}: {e}")
        else:
            print(f"Warning: Invalid y_score for {image_paths[i]}, max: {y_score_max}")

    mAP = np.mean(mAP_scores) if mAP_scores else 0.0
    print(f"{model_name} mAP: {mAP:.3f} ({len(mAP_scores)} valid queries)")

    if mAP_scores:
        precision, recall, _ = precision_recall_curve(y_true, y_score)
        plt.figure()
        plt.plot(recall, precision, label=f"{model_name} (mAP: {mAP:.3f})")
        plt.xlabel("Recall")
        plt.ylabel("Precision")
        plt.title(f"Precision-Recall Curve - {dataset_name}")
        plt.legend()
        plt.show()

        top_k = sim_idx[0, :5]
        bottom_k = sim_idx[0, -5:][::-1]
        print(f"{model_name} Top 5 retrieved:")
        for idx in top_k:
            print(f"  {image_paths[idx]} (Class: {labels[idx]})")
        print(f"{model_name} Worst 5 retrieved:")
        for idx in bottom_k:
            print(f"  {image_paths[idx]} (Class: {labels[idx]})")
        visualize_results(0, top_k, bottom_k, image_paths, labels, model_name)

# Main execution
for dataset_name, dataset_path in datasets.items():
    image_paths, labels = load_dataset(dataset_path)
    features_dict = {}

    for model_name, load_func in modelss.items():
        model, preprocess, feature_size = load_func()
        features = [extract_features(model, preprocess, path, feature_size) for path in image_paths]
        features_dict[model_name] = np.array(features)

    for model_name, features in features_dict.items():
        evaluate_model(features, labels, image_paths, model_name)
```
